```python
from PIL import Image
import numpy as np
from matplotlib import pyplot as plt
import cv2

#importing packages
```

```python
photo = Image.open("task1_3.jpg")
photo.size


#Open the image and save it as "photo"
```

```python
photo = photo.resize([180,270]) #resize因为原始图片比较大不适合展示
grey = photo.convert("L")
grey
#convert the photo to greyscale ie 黑白only
```

```python
grey= np.array(grey)/255.0
grey.shape
# 将jpg图片变成矩阵，并用0-1表示每个channel的深度
#看一下像素
```

```python
x,y= np.gradient(grey)
y.shape
plt.imshow(y)
#x，y分别储存了横向与纵向的一阶导 ie边缘
```

```python
Im2=np.abs(x)*255 #从0-1变回0-255 因为下面的image要以unit8的形式储存
newIm=Image.fromarray(Im2.astype("uint8"))
newIm
#从x中画出之前grey图片的横向边缘，背景色为黑
```

```python
Im3=(1-np.abs(x))*255
newIm=Image.fromarray(Im3.astype('uint8'))
newIm
#从x中画出之前grey图片的横向边缘，背景色为白（从黑图中减去黑色的部分）
```

```python
Im4=(1-np.abs(x)**0.25)*255 #想要保留黑色背景，只减去很小一部分的x
newIm=Image.fromarray(Im4.astype('uint8'))
newIm
```

可以看到效果并不是很好，有大量留白很粗糙。也许一般的代数运算无法解决这个问题？
